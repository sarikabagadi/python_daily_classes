```python
# compile time / method overloding polymorphism
class BCA_1St_semester :
  def student(self,a,b=0,c=0,d=0):
    return a+b+c+d

BCA =BCA_1St_semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,4,-12))


```

```python
# Run time / method overriding
class Animal :
  def sound(self):
    print("Animal makes a sound")

class Dog(Animal):
  def sound(self):
    print("BOW BOW BOW BOW")

class Cat(Animal):
  def sound(self):
    print("MEOW MEOW MEOW")
class tiger(Animal):
  def sound(self):
    print("ROAR ROAR ROAR")
class sneck(Animal):
  def sound(self):
    print("BUUUSSSSSS")
class cow(Animal):
  def sound(self):
    print("MOO MOO MOO")

obj1 = Animal()
obj1.sound()

obj2 = Dog()
obj2.sound()

obj3 = Cat()
obj3.sound()

obj4 = tiger()
obj4.sound()

obj5 = sneck()
obj5.sound()

obj6 = cow()
obj6.sound()
```

```python
#operator overloading polymorphism
class book:
  def __init__(self,pages):
    self.pages = pages
  def __add__(self,other):
    return self.pages + other.pages
obj1 = book(200)
obj2 = book(300)
print("Total pages is",obj1 + obj2)
```

```python
#Abstraction example
from abc import ABC,abstractmethod
class vehicle(ABC):
  def sound(self):
      pass
class car (vehicle):
  def sound(self):
    print("i'm driving a car")

obj = car()
obj.sound()
```

```python
# Abstraction examples
from abc import ABC,abstractmethod
class game(ABC):
  # normal function or method
  def free_fire (self):
    print("winner winner chicken dinner")

    # Abstract method/ function
  def BGMI(self):
    pass
    #normal function / method
  def Ludo_king(self):
    print("I'm win the game")
class temple_Run (game):

  #BGMI function implemented by sub class
  def BGMI(self):
    print("I got Royel pass Diamond")

    #normal function
  def chess(self):
    print("I'm district player")

obj = temple_Run()
obj.free_fire()
obj.BGMI()
obj.Ludo_king()
obj.chess()



```
